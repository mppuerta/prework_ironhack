```python
variables = ['rock', 'paper', 'scissors','lizard','spock']

maxgames = 0
wins = 0

def games():
    "Function that sets the maximum number of rounds"
    while True:
        global maxgames
        global wins
        try:
            maxgames = int(input("Select the maximum number of rounds: "))
            print('')
            oddnumber = maxgames/2
            wins = int(float(oddnumber)) + 1
        except:
            print("That's not even a number!")
            continue
        try:
            if oddnumber == int(float(oddnumber)):
                raise ValueError
            else:
                return maxgames
                break
        except:
            print("That's not an odd number :)")
    
    
comp = 0
hum = 0
battle = 0  
    
import random 
    
def computer():
    "Function that randomly returns a value"
    global comp
    comp = random.choice(variables)


def human():
    "Function that asks player for his choice"
    while True:
        global hum
        hum = input("Human's choice: ")
        if hum in variables:
            return hum
        else:
            print("I'm sorry, you can't battle with that... Try again!")

            
def battles(comp,hum):
    "Function that resolves the combact"
    global battle
    if hum == 'rock':
        if comp == 'rock':
            battle = 0
        elif comp == 'paper':
            battle = 1
        elif comp == 'scissors':
            battle = 2
        elif comp == 'lizard':
            battle = 2
        elif comp == 'spock':
            battle = 1
            
    elif hum == 'paper':
        if comp == 'rock':
            battle = 2
        elif comp == 'paper':
            battle = 0
        elif comp == 'scissors':
            battle = 1
        elif comp == 'lizard':
            battle = 1
        elif comp == 'spock':
            battle = 2
            
    elif hum == 'scissors':
        if comp == 'rock':
            battle = 1
        elif comp == 'paper':
            battle = 2
        elif comp == 'scissors':
            battle = 0
        elif comp == 'lizard':
            battle = 2
        elif comp == 'spock':
            battle = 1
      
    elif hum == 'lizard':
        if comp == 'rock':
            battle = 1
        elif comp == 'paper':
            battle = 2
        elif comp == 'scissors':
            battle = 1
        elif comp == 'lizard':
            battle = 0
        elif comp == 'spock':
            battle = 2
            
    elif hum == 'spock':
        if comp == 'rock':
            battle = 2
        elif comp == 'paper':
            battle = 1
        elif comp == 'scissors':
            battle = 2
        elif comp == 'lizard':
            battle = 1
        elif comp == 'spock':
            battle = 0

            
Human_won = 0
Machine_won = 0
            
            
def scoreboard():
    "Function that shows each player's choice and the state of the game"
    print("Machine's choice:", comp)
    global Human_won
    global Machine_won
    if battle == 2:
        Human_won += 1
    elif battle == 1:
        Machine_won += 1
    elif battle == 0:
        print('TIE')
    print("Human won:", Human_won)
    print("Machine won:", Machine_won)


games()


maxwins = 0

for i in range(maxgames):
    computer()
    human()
    battles(comp,hum)
    scoreboard()
    maxwins = max([Human_won,Machine_won])
    results = {'Human': Human_won, 'Machine': Machine_won}
    if maxwins == wins:
        for key,value in results.items():
            if value == wins:
                print('')
                print ('Winner: ', key)
                break
    elif i == (maxgames - 1):
        if Human_won == Machine_won:
            print('')
            print('There was a tie')
            break
        elif Human_won != Machine_won:
            winner = max(results.values())
            for key,value in results.items():
                if value == winner:
                    print('')
                    print ('Winner: ', key)
                    break
        
    print("")
```
